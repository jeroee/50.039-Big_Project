```python
'''
splits.ipynb file shows how the we split our dataset into 80-10-10 train validation test splits.

The data are also preprocessed by 
1) triming the videos to the first 400 frames
2) extracting 10 arousal and valence scores at intervals of 40 frames. 
eg: Arousal and valence score at frame 40, at frame 80, at frame 120 ... , at frame 400, 

All the preprocessed data are sent to the data folder.

>data
 > video
      > train_trim
      > valid_trim
      > test_trim
>label
 > arousal
      > train_trim
      > valid_trim
      > test_trim
 > valence
      > train_trim
      > valid_trim
      > test_trim
'''


import os 
import random
source_videos = r'C:\Users\jeroe\Documents\DeepLearning\Big_Project\aff_wild_annotations_bboxes_landmarks_new\videos\train'
source_valence = r'C:\Users\jeroe\Documents\DeepLearning\Big_Project\aff_wild_annotations_bboxes_landmarks_new\annotations\train\valence'
source_arousal = r'C:\Users\jeroe\Documents\DeepLearning\Big_Project\aff_wild_annotations_bboxes_landmarks_new\annotations\train\arousal'

video_train = r'C:\Users\jeroe\Documents\DeepLearning\Big_Project\data\video\train'
video_val = r'C:\Users\jeroe\Documents\DeepLearning\Big_Project\data\video\valid'
video_test = r'C:\Users\jeroe\Documents\DeepLearning\Big_Project\data\video\test'

valence_train = r'C:\Users\jeroe\Documents\DeepLearning\Big_Project\data\label\valence\train'
valence_val = r'C:\Users\jeroe\Documents\DeepLearning\Big_Project\data\label\valence\valid'
valence_test = r'C:\Users\jeroe\Documents\DeepLearning\Big_Project\data\label\valence\test'

arousal_train = r'C:\Users\jeroe\Documents\DeepLearning\Big_Project\data\label\arousal\train'
arousal_val = r'C:\Users\jeroe\Documents\DeepLearning\Big_Project\data\label\arousal\valid'
arousal_test = r'C:\Users\jeroe\Documents\DeepLearning\Big_Project\data\label\arousal\test'
```

```python
import cv2
import os
for i in os.listdir(video_train):
    cap = cv2.VideoCapture(os.path.join(video_train,i))
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    width  = cap.get(3)  # float `width`
    height = cap.get(4)  # float `height`
    i = i[:-4]+'.mp4'
    out = cv2.VideoWriter(os.path.join(video_trim_train,i), fourcc, 30.0, (int(width),int(height)))

    frame_count=0
    while frame_count<400:
        _,frame = cap.read()
        out.write(frame)
        frame_count+=1
    cap.release()
    out.release()
    cv2.destroyAllWindows()
```

```python
# trimming valence and arousal labels for 10 frames in intervals of 40 (40,80,120,160,200,240,280,320,360,400)
```

```python
for i in os.listdir(arousal_train):
    f_read = open(os.path.join(arousal_train,i), "r")
    lines = [ line.strip() for line in f_read ]
    if len(lines)<=400:
        print(i)
        continue
    else:
        f_write = open(os.path.join(arousal_trim_train,i),"w+")
        jump = 40
        for i in range(jump,400+jump,jump):
            #print(lines[i])
            f_write.write(lines[i]+'\n')
        f_write.close()
        f_read.close()
```
